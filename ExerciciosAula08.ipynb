## Exercício 1:

Resolução numérica de equações diferenciais. Imagine que tem um/a maratonista treinando, que num trecho de $5 km$ tem velocidade quase constante de $12 km/h$. Produza uma lista distância percorrida em intervalos de 1 minuto.


```
def cost_vel(di, df, v): # km e h
    d = df - di
    ds = []
    traveled = di
    while traveled < df:
        traveled += v
        ds.append(traveled)
    return ds

# Valor dividido por 60 para conversão de horas para minutos
y = cost_vel(0, 5, 12 / 60)
print(y)
```

## Exercício 2:

O/A mesmo/a atleta no seguinte trecho acelera por $200$ metros até chegar em $15 km/h$ por $2 km$. Queremos obter uma tabela ou gráfico da distância percorrida en função do tempo e o tempo total para chegar em $7 km$, faça isto utilizando listas.


```
import matplotlib.pyplot as plt

def aceleration(di, df, vi, vf): # km e h
    d = df - di
    v = vf - vi
    ds = []
    traveled = di
    while traveled < df:
        traveled += vi + d * v
        ds.append(traveled)
    return ds

def cost_vel(di, df, v): # km e h
    d = df - di
    ds = []
    traveled = di
    while traveled < df:
        traveled += v
        ds.append(traveled)
    return ds

def minutes(distances):
    ts = []
    for i in range(len(distances)):
        ts.append(i)
    return ts

# Valores divididos por 60 para conversão de horas para minutos
a = cost_vel(0, 5, 12 / 60)
b = aceleration(5, 5.2, 12 / 60, 15 / 60)
c = cost_vel(5.2, 7, 15 / 60)

y = [0]
y.extend(a)
y.extend(b)
y.extend(c)

x = minutes(y)

print(y)
print(x)

plt.plot(x, y)
plt.xlabel('Tempo (min)')
plt.ylabel('Distância (km)')
```

## Exercício 3:

Considere um péndulo simples num plano. O comprimento da linha é $1$ metro, a massa é $0,2 kg$ e a aceleração de gravidade $9,8 m/s^2$. Faça um gráfico ou crie uma tabela para a posição em coordenadas cartesianas em função do tempo, utilizando listas.


```
from numpy import arange, radians, sqrt, cos
import matplotlib.pyplot as plt

# Plot:
def plot_results(time, theta):
    plt.plot(time, theta)
    plt.title('Simple pendulum   ->   Initial angle = ' + str(initial_angle) + '°')
    plt.xlabel('time (s)')
    plt.ylabel('angle (rad)')
    plt.grid(True)
    plt.show()

# Pendulum - Linear problem:
def pendulum(g, l, time, initial_angle):
    w = sqrt(g / l)
    theta0 = radians(initial_angle)
    return [theta0 * cos(w * t) for t in time]

# Parameters:
g = 9.8
l = 1
time = arange(0, 10, 0.025)
initial_angle = 10

# Results:
theta = pendulum(g, l, time, initial_angle)
plot_results(time, theta)
```
