## Exercício 1:

Escreva um conjunto apropriadamente geral de funções que possam desenhar flores.


```
from math import pi

def polyline(t, n, length, angle):
  for i in range(n):
    t.forward(length)
    t.left(angle)

def arc(t, r, angle):
  arc_length = 2 * pi * r * abs(angle) / 360
  n = int(arc_length / 4) + 3
  step_length = arc_length / n
  step_angle = float(angle) / n

  t.left(step_angle / 2)
  polyline(t, n, step_length, step_angle)
  t.right(step_angle / 2)

def petal(t, r, angle):
  for i in range(2):
    arc(t, r, angle)
    t.left(180 - angle)

def flower(t, n, r, angle):
  for i in range(n):
    petal(t, r, angle)
    t.left(360.0 / n)

def move(t, distance):
  t.penup()
  t.right(90)
  t.forward(distance)
  t.left(90)
  t.pendown()

window = Turtle
window.initializeTurtle()
window.speed(13)
window.width(1)

move(window, -200)
flower(window, 7, 60, 60)
move(window, 200)
flower(window, 10, 40, 80)
move(window, 200)
flower(window, 20, 140, 20)
window.hideturtle()
```

## Exercício 2:

Escreva um conjunto apropriadamente geral de funções que podem desenhar formas.


```
from math import sin, pi

def isosceles(t, r, angle):
  y = r * sin(angle * pi / 180)
  t.right(angle)
  t.forward(r)
  t.left(90 + angle)
  t.forward(2 * y)
  t.left(90 + angle)
  t.forward(r)
  t.left(180 - angle)

def pie(t, r, n):
  angle = int(180 / n)
  for i in range(n):
    t.right(angle * 2)
    isosceles(t, r, angle)

window = Turtle
window.initializeTurtle()
window.speed(7)
window.width(1)

move(window, -220)
pie(window, 100, 5)
move(window, 210)
pie(window, 100, 6)
move(window, 210)
pie(window, 100, 10)
window.hideturtle()
```

## Exercício 3:

Adicione instruções para desenhar o talo e folha da flor. Faça a refatoração apropriada para a interface da flor completa ficar "limpa".


```
def moveUp(t, distance):
  t.penup()
  t.forward(distance)
  t.pendown()

def flowerDisk(t, r):
  move(t, r)
  arc(t, r, 360)
  move(t, -r)

def stem(t, width, height):
  move(t, -(width / 2))
  for i in range(2):
    t.right(90)
    t.forward(width)
    t.right(90)
    t.forward(height)
  move(t, (width / 2))

def leafs(t, r, angle):
  t.left(45)
  petal(t, r, angle)
  t.left(180)
  petal(t, r, angle)
  t.right(225)

window = Turtle
window.initializeTurtle()
window.speed(13)
window.width(1)

moveUp(window, 100)
flower(window, 10, 40, 80)
flowerDisk(window, 10)
stem(window, 6, 200)
moveUp(window, -100)
leafs(window, 30, 90)
window.hideturtle()
```

```
from math import pi

def archimedeanSpiral(t, r, loops):
  loops *= 360
  arc_length = 2 * pi * r * abs(loops) / 360
  n = int(arc_length / 4) + 3
  step_length = arc_length / n
  step_angle = float(loops) / n

  t.right(step_angle / 2)
  for i in range(n):
    t.forward(step_length + i)
    t.right(step_angle)
  t.left(step_angle / 2)

window = Turtle
window.initializeTurtle()
window.speed(13)

archimedeanSpiral(window, 13, 4)
```
