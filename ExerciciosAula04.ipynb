## Exercício 1:

Escreva uma função com um valor de retorno que:
* tome como dois argumentos;
* execute a soma dos dois parâmetros que foram passados como argumento e atribua o resultado a uma variável;
* tenha como valor de retorno o dobro da soma dos dois parâmetros.


```
def somar(x, y):
  return x + y

def dobrar(valor):
  return 2 * valor

soma = somar(4, 6)
dobroDaSoma = dobrar(soma)

print(dobroDaSoma)
```

## Exercício 2:

Escreva uma versão mais geral de print_spam, chamada print_twice, que use uma string como parâmetro e imprima duas vezes. Use a versão modificada de do_twice para chamar print_twice duas vezes, passando 'spam' como um argumento. Defina uma nova função chamada do_four que recebe um objeto de função e um valor e chama a função quatro vezes, passando o valor como um parâmetro. Deve haver apenas duas declarações no corpo desta função, não quatro.


```
def print_twice(valor):
  print(valor)
  print(valor)

def do_twice(valor):
  print_twice(valor)
  print_twice(valor)

do_twice('Bla!')
```
