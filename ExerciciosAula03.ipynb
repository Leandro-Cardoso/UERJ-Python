## Exercício 2:

Crie uma função que calcule e imprima velocidade media de um objeto a partir de uma posição inicial, a final e o tempo transcorrido para um objeto em MRU. Também crie uma funcão que calcule e imprima a velocidade de um objeto a partir da aceleração constante e o tempo (MRUA) (p.ex. queda libre).

```
def velocidadeMedia(posicaoInicial, posicaoFinal, tempo):
  print('A velocidade média é de', (posicaoFinal - posicaoInicial) / tempo)

def velocidadeAceleracao(aceleracao, tempo):
  print('A velocidade a partir da aceleração constante é de', aceleracao * tempo)

velocidadeMedia(10, 40, 2)
velocidadeAceleracao(10, 5)
```

## Exercício 3:

Crie uma funcão para calcular o ángulo zenital do sol (da semana passada) tomando como argumento as medidas da altura e o comprimento da sombra.

```
from math import atan, degrees

def anguloZenital(altura, comprimento):
  print('O ângulo é de %.2f°' % degrees(atan(altura / comprimento)))

anguloZenital(1.7, 3.1)
```

## Exercício 4:

Crie uma função que faça a conversão de uma medida inicialmente em milhas para m, e outra para o inverso; uma de horas para segundos, e o inverso. Utilize estas funções para resolver novamente o primeiro exercício da semana passada (da corrida). Se uma pessoa demora 30 minutos em 4 milhas, qual velocidade media em km/h ? e o tempo medio por kilometro?

```
def milhasParaMetros(milhas):
  return milhas * 1609.344

def metrosParaMilhas(metros):
  return metros / 1609.344

def horasParaSegundos(horas):
  return horas * 60 * 60

def segundosParaHoras(segundos):
  return segundos / 60 / 60

print('A conversão de 5 milhas para metros:', milhasParaMetros(5))
print('A conversão de 2000 metros para milhas:', metrosParaMilhas(2000))
print('A conversão de 2 horas para segundos:', horasParaSegundos(2))
print('A conversão de 4000 segundos para horas:', segundosParaHoras(4000))

tempo = 0.5 # EM HORAS
distancia = milhasParaMetros(4) / 1000 # EM KM
velocidade = distancia / tempo # KM/H
tempoPorKm = tempo / distancia # H/KM

print('\nA velocidade média é de %.2f Km/h e o tempo médio por Km é de %.2f segundos' % (velocidade, horasParaSegundos(tempoPorKm)))
```

## Exercício 5:

Crie funções para calcular os outros exemplos das aulas anteriores: IMC, volume de uma esfera, distancia entre pontos de máximos de difração. Decida quais serão os argumentos e o valor retornado.

```
from math import pi

def imc(peso, altura): # Kg e metros
  return peso / altura ** 2

def volumeEsfera(raio):
  return 4 / 3 * pi * raio ** 3

def distanciaMaximosDifracao(comprimento, distancia, espacamento):
  return comprimento * distancia / espacamento

print('O IMC de uma pessoa com 80 Kg e 1.7 m é de:', imc(80, 1.7))
print('\nO volume de uma esfera com 5 m de raio é de:', volumeEsfera(5))
print('\nA distancia entre os pontos de máximos de difração em mm é de:', distanciaMaximosDifracao(632.8 * 10 ** -6, 1.98 * 10 ** 3, 0.25))
```
