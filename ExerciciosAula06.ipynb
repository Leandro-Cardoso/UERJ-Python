## Exercício 2:

Repita o exercício 1 usando Manipulação de exceção em Python (pesquisem), com as cláusulas try e except.


```
import ColabTurtle.Turtle as tess

def draw_bar(t, height, width):
    """ Get turtle t to draw one bar, of height. """
    t.forward(int(height))
    t.right(90)
    t.forward(width)
    t.right(90)
    t.forward(int(height))
    t.left(180)

def is_number(value):
    try:
         int(value)
    except ValueError:
         return False
    return True

def create_list():
    items = []
    print('Escreva numero inteiro referente ao item da lista ou digite "fim" para terminar a lista.')
    while True:
        item = input('Item = ')
        if item == 'fim' or item == 'Fim':
            return items
        elif not (is_number(item)):
            print('Digite apenas numeros inteiros.')
        else:
            items.append(item)

def start_position(t, items, bar_width):
    move = int(len(items) * bar_width / 2)
    t.penup()
    t.left(90)
    t.forward(move)
    t.left(90)
    t.forward(200)
    t.right(180)
    t.pendown()

xs = create_list()
bar_width = 50

tess.initializeTurtle()   # inicialização

tess.color("blue")
tess.bgcolor("pink")
tess.speed(8)
tess.pensize(2)

start_position(tess, xs, bar_width)

for a in xs:
    draw_bar(tess, a, bar_width)

tess.hideturtle()
```

## Exercício 3:

Faça um programa equivalente ao do exercício 1 utilizando matplotlib e numpy ao invés de turtle.


```
import numpy as np
import matplotlib.pyplot as plt

def is_int(value):
    try:
         int(value)
    except ValueError:
         return False
    return True

def set_ages():
    items = []
    print('Escreva idade dos alunos de uma turma ou digite "fim" para terminar a lista.')
    while True:
        item = input('Item = ')
        if item == 'fim' or item == 'Fim':
            return np.array(items)
        elif is_int(item):
            items.append(abs(int(item)))
        else:
            print('Digite apenas numeros inteiros.')

def set_bins():
    print('Qual a quantidade de intervalo de valores do histograma? (bins)')
    while True:
        bins = input('Bins = ')
        if is_int(bins):
            return int(bins)
        else:
            print('Digite apenas numeros inteiros.')

ages = set_ages()
bins = set_bins()

plt.figure(figsize = (10, 6))
plt.title('Frequência x Idade')
plt.hist(ages, bins = bins, label = 'Frequência x Idade', color = 'red')

print(ages)
print(bins)
```
