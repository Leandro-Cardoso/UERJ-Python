## Exercício 1:
Use o Python como uma calculadora. E se você quiser calcular o seno de um dado ângulo? Consegue fazer isso com Python?

```
from math import sin

angulo = 60
seno = sin(angulo)

print('O seno do ângulo de %s° é %.4f' % (angulo, seno))
```

## Exercício 2:
Faça mais alguns exemplos com os operadores de comparação, principalmente misturando expressões antes e depois dos operadores.

```
from math import sin, cos

angulo = 90
seno = sin(angulo)
cosseno = cos(angulo)
comparacao = 'maior que ' if seno > cosseno else 'igual a' if seno == cosseno else 'menor que '

print('O seno do ângulo de %s° é %.4f' % (angulo, seno))
print('O cosseno do ângulo de %s° é %.4f' % (angulo, cosseno))
print('\nO seno é %so cosseno' % comparacao)
```
