## Exercício 1:

A conjectura de Collatz, diz que "Todos os inteiros positivos irão eventualmente convergir para 1 usando as regras do Collatz". Faça um programa que verifique a conjectura de Collatz para alguns valores. As regras de Collatz são: Dado um número n:

* se n for par, divida-o por 2;
* Se n for ímpar, atualize-o para 3*n + 1


```
def collatz(n):
    if n % 2 == 0:
        n /= 2
    else:
        n = 3 * n + 1
    return n

def conjectura_collatz(n):
    if n > 0 and isinstance(n, int):
        ns = [n]
        while n != 1:
            n = collatz(n)
            ns.append(int(n))
        print(ns)
    else:
        print('O número precisa ser inteiro e positivo.')

conjectura_collatz(10)
conjectura_collatz(0)
conjectura_collatz(-15)
conjectura_collatz(5.35)
conjectura_collatz(15)
```

## Exercício 2:

Modifique a função counter para contar o número de vezes que os dígitos 0 e 5 aparecem em um dado número n. Por exemplo, n = 10568, tem 2 dígitos 0 ou 5.


```
def counter(n):
    count = 0
    while n!=0:
        count = count + 1
        n = n // 10         # divisão de inteiros
    return count

def counter_0_5(n):
    n = str(n)
    count = 0
    for i in n:
        if i == '0' or i == '5':
            count += 1
    return count

print(counter(10568))
print(counter_0_5(10568))
```

## Exercício 4:

O método de Newton para raiz quadrada. Suponha que você queira calcular a raiz quadrada de n. Se você começar com quase qualquer aproximação, poderá calcular uma aproximação melhor (mais próxima da resposta real) com a seguinte fórmula: melhor = (approx + n/approx)/2

Usando um loop e repetindo esta fórmula até que a melhor aproximação se aproxime o suficiente da anterior (podemos considerar uma diferença de 0,001), podemos escrever uma função para calcular a raiz quadrada.

Esse é um exemplo de repetição indefinida. Com qual tipo de loop você deve escrever o seu programa?


```
def newton_sqr(x, accuracy = 0.001):
    if x < 0:
        return 'Insira somente valores positivos.'
    approx = x
    while approx ** 2 > x + accuracy or approx ** 2 < x - accuracy:
        approx = (approx + x / approx) / 2
    return approx

print(newton_sqr(-5))
print(newton_sqr(0))
print(newton_sqr(0.5))
print(newton_sqr(1))
print(newton_sqr(2))
print(newton_sqr(3))
print(newton_sqr(4))
print(newton_sqr(9785))
```

O módulo time fornece uma função, também chamada time, que devolve a Hora Média de Greenwich na “época”, que é um momento arbitrário usado como ponto de referência. Em sistemas UNIX, a época é primeiro de janeiro de 1970.
Escreva um script que leia a hora atual e a converta em um tempo em horas, minutos e segundos, mais o número de dias desde a época.

```
from time import time, localtime

def seconds_to_days(s):
    return (s / 60 / 60 / 24)

def draw_days():
    print(seconds_to_days(time()), 'dias passados, desde a época de referência.')

def draw_time(timezone = -3, dst = 0):
    t = localtime()
    hour = t[3] + timezone + dst
    if hour < 0:
        hour += 24
    elif hour > 23:
        hour -= 24
    print('Horário atual:', hour, ':', t[4], ':', t[5])

draw_time()
draw_days()
```

O último teorema de Fermat diz que não existem números inteiros a, b e c tais que a^n + b^n == c^n para quaisquer valores de n maiores que 2.

* Escreva uma função chamada check_fermat que receba quatro parâmetros – a, b, c e n – e verifique se o teorema de Fermat se mantém. Se n for maior que 2 e a^n + b^n == c^n o programa deve imprimir, “Holy smokes, Fermat was wrong!” Senão o programa deve exibir “No, that doesn’t work.”

* Escreva uma função que peça ao usuário para digitar valores para a, b, c e n, os converta em números inteiros e use check_fermat para verificar se violam o teorema de Fermat.


```
def check_fermat(a, b, c, n):
    if n > 2 and (a ** n) + (b ** n) == (c ** n):
        print('Holy smokes, Fermat was wrong!')
    else:
        print("No, that doesn't work.")
    
def set_int_value(xvar):
    while True:
        x = input(str(xvar) + ' = ')
        try:
            x = int(x)
            return int(x)
        except:
            print('Digite apenas um número inteiro.')

def set_abcn():
    print('Digite apenas números inteiros:')
    a = set_int_value('a')
    b = set_int_value('b')
    c = set_int_value('c')
    n = set_int_value('n')
    check_fermat(a, b, c, n)

set_abcn()
```

Senão, você pode. (Se a soma de dois comprimentos igualar o terceiro, eles formam um triângulo chamado “degenerado”.)

* Escreva uma função chamada is_triangle que receba três números inteiros como argumentos, e que imprima “Yes” ou “No”, dependendo da possibilidade de formar ou não um triângulo de gravetos com os comprimentos dados.
* Escreva uma função que peça ao usuário para digitar três comprimentos de gravetos, os converta em números inteiros e use is_triangle para verificar se os gravetos com os comprimentos dados podem formar um triângulo.


```
def is_triangle(a, b, c):
    higher_number = max([a, b, c])
    if (a == higher_number and a <= b + c) or (b == higher_number and b <= a + c) or (c == higher_number and c <= a + b):
        print('Yes')
    else:
        print('No')

def set_int_positive_value(xvar):
    while True:
        x = input(str(xvar) + ' = ')
        try:
            x = int(x)
            if x < 0:
                print('O valor precisa ser maior que 0.')
            else:
                return int(x)
        except:
            print('Digite apenas um número inteiro.')

def set_abc():
    print('Digite apenas números inteiros:')
    a = set_int_positive_value('a')
    b = set_int_positive_value('b')
    c = set_int_positive_value('c')
    is_triangle(a, b, c)

set_abc()
```

* Escreva uma função chamada koch que receba um ColabTurtle (ou turtle) e um comprimento como parâmetros, e use o ColabTurtle para desenhar uma curva de Koch com o comprimento dado.

* Escreva uma função chamada snowflake que desenhe três curvas de Koch para fazer o traçado de um floco de neve.

* A curva de Koch pode ser generalizada de vários modos. Veja exemplos em http://en.wikipedia.org/wiki/Koch_snowflake e implemente o seu favorito.

```
def start_position(t, size):
  t.penup()
  t.forward(size / 3.5)
  t.right(90)
  t.backward(size / 2)
  t.pendown()

def koch(t, size):
    if size < 10:
        t.forward(size)
        return
    fraction_size = size / 3
    koch(t, fraction_size)
    t.left(60)
    koch(t, fraction_size)
    t.right(120)
    koch(t, fraction_size)
    t.left(60)
    koch(t, fraction_size)

def snowflake(t, size):
    start_position(t, size)
    for i in range(3):
        koch(t, size)
        t.right(120)

t = Turtle
t.initializeTurtle()
t.speed(13)
t.width(1)
t.hideturtle()
snowflake(t, 400)
```
