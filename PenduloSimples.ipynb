# Pendulo Simples

### Vetores

```
from numpy import sin, cos, radians, sqrt, arange
from scipy.integrate import odeint

# functions:
def linear_pend():
    ''' Retorna uma lista em radianos com a solução linear. '''
    w = sqrt(g / l)
    return [theta0 * cos(w * t) for t in time]

def nonlinear_equation(y0, t):
    ''' Função complementar para a solução não linear. '''
    theta, x = y0
    return [x, -(g / l) * sin(theta)]

def  nonlinear_pend():
    ''' Retorna uma lista em radianos com a solução não linear. '''
    return odeint(nonlinear_equation, [theta0, x0], time)

# Parameters:
g = 9.8
l = 1
time = arange(0, 10, 0.025)
initial_angle = 30

theta0 = radians(initial_angle)
x0 = radians(0.0)

# Results:
nonlinear = nonlinear_pend()
linear = linear_pend()

print('Solução não linear:', nonlinear)
print('Solução linear:', linear)
```

### Gráfico


```
import matplotlib.pyplot as plt

# functions:
def plot_results(time, theta1, theta2):
    ''' Gera um gráfico para comparação entre a solução linear e a não linear. '''
    plt.plot(time, theta1[:, 0])
    plt.plot(time, theta2)
    plt.title('Simple pendulum   ->   Initial angle = ' + str(initial_angle) + '°')
    plt.xlabel('time (s)')
    plt.ylabel('angle (rad)')
    plt.grid(True)
    plt.legend(['Nonlinear', 'Linear'], loc = 'lower right')
    plt.show()

# Results:
plot_results(time, nonlinear, linear)
```
